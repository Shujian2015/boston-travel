```python
import pandas as pd
import json

df = pd.read_csv('reviews.csv')

d = {}
for _, row in df.iterrows():
    key = row['listing_id'] 
    if key in d:
        d[key] += str(row['comments'])
    else:
        d[key] = str(row['comments'])
        
# print(d)        
```

```python
import rake
import operator

rake_object = rake.Rake("SmartStoplist.txt", 3,3,1)
results = {}

for k, v in dict.iteritems(d):
    keywords = rake_object.run(str(v))
    kw = [x[0] for x in keywords[:5]]
    results[k] = ', '.join(kw)
```

```python
print(results)
```

```python
import json
json.dump(results, open("keywords.txt",'w'))

print('Done')
```
